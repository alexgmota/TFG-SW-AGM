## Data overview.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
import cv2
from glob import glob
```

```python
imgs = []
imgIdx = random.randint(80, 159)
paths = glob(f'../*/video01/video01_00080/frame_{imgIdx}*')
for sample in paths:
    img = cv2.imread(sample)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imgs.append(img)
imgs = np.array(imgs)

plt.figure(figsize=(20,20))
for i in range(imgs.shape[0]):
  plt.subplot(141 + i)
  plt.imshow(imgs[i])
  plt.title(paths[i].split('\\')[-1])
  plt.axis("off")
plt.tight_layout()
plt.show()
```
