```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
span = 100
lgnd = np.zeros((13*100, 800))
for i in range(13):
    lgnd[i*span:(i+1)*span] = i
```

```python
labels_encoding = {
    0: 'BG',
    1: 'Abdominal Wall',
    2: 'Liver',
    3: 'Gastrointestinal Tract',
    4: 'Fat',
    5: 'Grasper',
    6: 'Connective Tissue',
    7: 'Blood',
    8: 'Cystic Duct',
    9: 'L-hook Electrocautery',
    10: 'Gallbladder',
    11: 'Hepatic Vein',
    12: 'Liver Ligament',
}
colors = ['w', '#E0C202', '#E0D302', '#C6E002', '#7FE002', '#36E002', '#05E002', '#02E020', '#02E048', '#02E0AD', '#02CDE0', '#0277E0', '#7202E0']
```

```python
plt.figure()
plt.title('Legend')
plt.imshow(lgnd, cmap='magma', norm=plt.Normalize(vmin=0, vmax=12))
for i in range(13):
    plt.text(400, i*span + span/2, f"{list(labels_encoding.values())[i]}",
        ha="center", va="center", color=colors[i], fontsize='medium', )
plt.axis("off")
plt.show()
```
